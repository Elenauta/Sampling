### Exercici 1
Agafa un conjunt de dades de tema esportiu que t'agradi. Realitza un mostreig de les dades generant una mostra aleatòria simple i una mostra sistemàtica.

I'll use a storical dataset on the Modern Olympic Games, including all the Games from Athens 1896 to Rio 2016. 'athlete_events.csv', with a Public Domain License, from: https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results/version/2 ... I´ll focus on gender and age issues. 

```python
import numpy as np
import pandas as pd
import random
```

```python
# Importing dataset
df = pd.read_csv('athlete_events.csv')
df.head()
```

```python
# Drop some not needed columns
df.drop(['ID', 'Height', 'Weight', 'NOC', 'Games', 'City', 'Event'], axis=1, inplace=True)
df
```

```python
# Substituing NaN by ceros and checking it
df['Medal'] = df['Medal'].replace(np.nan, 0)
df['Medal'].unique()
```

```python
df.head(3)
```

```python
df.info()
```

```python
# Age atribute has some null entries
df.describe()
```

```python
# Filling this nulls whit the age mean
df['Age'] = df['Age'].fillna(25.5)
df.info()
```

```python
df.describe()
```

```python
# Who are this 10 and 97 year old?, wich sport and country they were representing? 
df.sort_values(['Age'], ascending=True)
```

```python
# Dimitrios Loundras, 10 year old, for Ethnikos Gymnastikos Syllogos in 1896, Summer	Gymnastics	Bronze
# John Quincy Adams Ward, 97 year old, for United States in 1928, Summer	Art Competitions	0
```

```python
# Explorem els atributs categorics
df = df.astype(dict((col, 'category') for col in df.columns if col not in ['ID', 'Name'] and df[col].dtype != float))
df.describe(include='category')
```

### Mostra aleatoria simple

```python
# The entire population has equal probabilities on been chosen
# I take a 10 percent sample
population_size = len(df)
sample_size = int(population_size * 0.01)

np.random.seed(0) # for same random results 
df_simple_sample = df.sample(sample_size)
df_simple_sample.head(3)
```

```python
df_simple_sample.describe()
```

```python
df_simple_sample.describe(include='category')
```

### Mostra sistemàtica

```python
# Here the sample is taken at regular intervals
k = int(population_size / sample_size)
random_first = np.random.randint(0, k)
df_sistem_sample = df[random_first : population_size : k]
df_sistem_sample.head(3)
```

```python
df_sistem_sample.describe()
```

```python
df_sistem_sample.describe(include='category')
```

Continua amb el conjunt de dades de tema esportiu i genera una mostra estratificada i una mostra utilitzant SMOTE (Synthetic Minority Oversampling Technique).

### Mostra estratificada

```python
# sample is taken after classifying
df.Sex.value_counts()
```

```python
from sklearn.model_selection import train_test_split

_, df_strat = train_test_split(df, stratify=df[['Sex']], test_size=0.01, random_state=0)
```

```python
df_strat.Sex.value_counts()
```

```python
df_strat.describe()
```

```python
df_strat.describe(include='category')
```
